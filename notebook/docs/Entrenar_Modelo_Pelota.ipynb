```python
import cv2
import os
from google.colab import drive

# 1. Montar Drive
drive.mount('/content/drive')

src_video = '/content/drive/MyDrive/Pickleball_Colab/partido1.mp4'
local_video = '/content/partido1_convertido.mp4'

# 2. Convertir el video a H.264 (formato universal para OpenCV)
print("⏳ Convirtiendo video a formato compatible...")
!ffmpeg -i {src_video} -c:v libx264 -preset fast -c:a copy -y {local_video}

if not os.path.exists(local_video):
    print("❌ Error al convertir el video.")
else:
    print(f"✅ Video convertido correctamente en entorno local.")

    # 3. Preparar carpeta de salida
    out_dir = '/content/frames_dataset'
    if os.path.exists(out_dir):
        import shutil
        shutil.rmtree(out_dir)
    os.makedirs(out_dir)

    # 4. Extraer frames del video LOCAL convertido
    cap = cv2.VideoCapture(local_video)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    fps = int(cap.get(cv2.CAP_PROP_FPS))

    print(f"📹 Video local: {total_frames} frames a {fps} FPS.")

    # Extraer 1 frame cada 10 frames
    extract_rate = 10
    count = 0
    saved = 0

    print(f"\n⏳ Extrayendo frames...")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        if count % extract_rate == 0:
            filename = os.path.join(out_dir, f'frame_{saved:04d}.jpg')
            cv2.imwrite(filename, frame)
            saved += 1
        count += 1

    cap.release()
    print(f"\n✅ {saved} frames extraídos y guardados en {out_dir}")
```
